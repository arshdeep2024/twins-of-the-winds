```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import make_scorer, mean_squared_error#, neg_mean_squared_error
np.random.seed(42)

data = pd.read_csv("train.csv")

#data_with_h = data
#data_without_h = data.drop("humidity", axis=1)
X1 = data.drop(["s.s.temp."], axis=1)
y1 = data["s.s.temp."]

X1.humidity.fillna(value=np.mean(X1.humidity), inplace=True)
X1["zon.winds"].fillna(value=np.mean(X1["zon.winds"]), inplace=True)
X1["mer.winds"].fillna(value=np.mean(X1["mer.winds"]), inplace=True)
X1["air temp."].fillna(value=np.mean(X1["air temp."]), inplace=True)

clf = RandomForestRegressor(n_jobs=-1)

X_train, X_test, y_train, y_test = train_test_split(X1, y1, test_size=0.09)
#Making grid of hyperparameters for GridSearchCV
grid = {
    'n_estimators': [200, 300],
    'max_depth': [None, 10],
    'min_samples_split': [5, 7],
    'min_samples_leaf': [1, 2]
}

scorer=make_scorer(mean_squared_error, greater_is_better=False)
# Doing Grid Search
gs_clf = GridSearchCV(estimator=clf,
                      param_grid=grid,
                      cv=7,
                      scoring=scorer,
                      n_jobs=-1,
                      verbose=2)

gs_clf.fit(X_train, y_train)
gs_clf.score(X_test, y_test)
```

```python
#print(X1)
testing_labels = pd.read_csv("data_1997_1998.csv")
testing_labels = testing_labels.drop(['Index'], axis=1)
testing_labels.humidity.fillna(value=np.mean(testing_labels.humidity), inplace = True)
testing_labels["zon.winds"].fillna(value=np.mean(testing_labels["zon.winds"]), inplace = True)
testing_labels["mer.winds"].fillna(value=np.mean(testing_labels["mer.winds"]), inplace = True)
testing_labels["air temp."].fillna(value=np.mean(testing_labels["air temp."]), inplace = True)
#print(testing_labels)
y_prediction_1997_1998 = gs_clf.predict(testing_labels)
y_prediction_1997_1998 = y_prediction_1997_1998.round(decimals=2)
y_prediction_1997_1998 = pd.Series(y_prediction_1997_1998)
y_prediction_1997_1998.to_csv("s.s.temp._1997_1998.csv", index=False)
y_prediction_1997_1998.head()
```
